# Machine Learning

Der Prozess, mit einem unterliegenden Datenset einen Algorithmus (das Model) zu erzeugen

In dem Datenset befinden sich Zusammenhänge; der Algorithmus soll diese Zusammenhänge selbstständig erkennen und speichern

Model: Fertiges Programm, welches beim ML Prozess erzeugt wird, und die Aussagen zu neuen Daten treffen kann

Beispiel Wetter-Modell:

Inputs: Datum, Uhrzeit, Wolken%, Niederschlag%

Output: Temperatur

20.12.2024, 14:00, 100%, 0% -> 2°C, 5°C, 7°C, 6°C; Realität: 6°C

Bei 4 Parametern kann dieser Algorithmus noch per Hand geschrieben werden, bei 10+ Parametern ist dies schon schwieriger

Wie funktioniert ChatGPT?

Diese Begriffe haben Zusammenhänge mit anderen Begriffen

Die Begriffe werden weiterverfolgt nach weiteren Begriffen

Diese Begriffe werden am Ende zu einem lesbaren Text zusammengebaut

### Aufteilung der Daten

Für den Trainingsprozess werden Trainingsdaten und nach dem Trainingsprozess Testdaten benötigt

Hierfür wird das originale datenset in 80% Trainingsdaten und 20% Testdaten aufgeteilt

```python
t = int(len(data) * 0.8)
```

```python
t
```

```python
len(data) - t
```

```python
data.sample()  # sample: Ein zufälliger Datensatz
```

```python
random = data.sample(frac=1)  # Mit frac=1: Datenset mischen
```

```python
training = random.iloc[0:t]
```

```python
test = random.iloc[t:]
```

```python
len(training)
```

```python
len(test)
```

### Unebenheiten der Klasse beheben

Wenn im Datenset eine ungleiche Menge an Daten vorhanden ist, kann das Modell eine Neigung (Bias) entwickeln

```python
training[14].value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler()
```

```python
left, right = ros.fit_resample(training.iloc[:, :-1], training.iloc[:, -1])  # 2 Parameter: Daten ohne class, class alleine
```

```python
training = pd.DataFrame(np.hstack((left, right.values.reshape(-1, 1))))
```

```python
training[14].value_counts()
```

```python
training = training.sample(frac=1)
```

#### Aufteilung Links/Rechts

```python
trainingL = training.iloc[:, :-1]
trainingR = training.iloc[:, -1]

testL = test.iloc[:, :-1]
testR = test.iloc[:, -1]
```

## Simple Algorithmen

### kNN

k-nearest neighbors

- Neuer Datenpunkt ohne Klasse wird platziert
- Jetzt wird geprüft, wieviele Punkte von welcher Klasse näher an dem Punkt sind
- Die Klasse die mehr Punkte in der nähe hat, gibt die Klasse an den neuen Punkt weiter

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(7)  # k = 7
```

```python
knn_model = knn.fit(trainingL, trainingR)
```

```python
knn_model
```

```python
predictions = knn_model.predict(testL)  # Vorhersagen aus dem Model mittels predict generieren lassen
```

```python
predictions
```

```python
testR.values
```

```python
predictions == testR.values
```

```python
pd.Series(predictions == testR.values).value_counts()
```

```python
def evaluate(prediction):
    counts = pd.Series(predictions == testR.values).value_counts()
    print(f"Korrekte Vorhersagen: {counts[True] / len(testR) * 100}%")
    print(f"Falsche Vorhersagen: {counts[False] / len(testR) * 100}%")
```

```python
evaluate(predictions)
```

```python
pd.DataFrame(np.hstack((testL, testR.values.reshape(-1, 1), predictions.reshape(-1, 1))))
```

### Naive Bayes

Verwendet Wahrscheinlichkeiten, um die Datenpunkte zu klassifizieren

Wenn ein Datenpunkt über 50% ist, wird er der einen Klasse zugewiesen, sonst der anderen

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
```

```python
nb_model = nb.fit(trainingL, trainingR)
```

```python
predictions = nb_model.predict(testL)
```

```python
evaluate(predictions)
```

### Logistische Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr_model = lr.fit(trainingL, trainingR)
```

```python
predictions = lr_model.predict(testL)
```

```python
evaluate(predictions)
```

### Support Vector Machines

Legt eine Hyperplane (H) zwischen alle Datenpunkte (mittig) und breitet im Anschluss ein Margin (M) aus

Alle Punkte die Links liegen werden zu Klasse 1 klassfiziert, alle anderen Punkte zur anderen Klasse

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc_model = svc.fit(trainingL, trainingR)
```

```python
predictions = svc_model.predict(testL)
```

```python
evaluate(predictions)
```

## Neurales Netzwerk

Besteht aus Knotenpunkten (Nodes), diese Nodes werden in Schichten angelegt und miteinander verbunden

Bei jedem Lernprozess werden die Daten durch dieses Netzwerk gesendet

Jede Node besteht aus einem Neuron

Neuron:
- Am Anfang werden die Inputs multipliziert und im Anschluss aufsummiert
- Danach wird diese Summe in eine Activation Function gesendet
- Activation Function: Normale Mathematische Funktion
- Am Ende des Neurons werden die Outputdaten an die nächste Schicht weitergegeben

Am Ende eines Durchgangs fängt der Lernprozess von vorne an

```python
import tensorflow as tf
```

```python
# keras: Tensorflow API, erleichtert das Erstellen von Modellen
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(14,)),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])
```

```python
model.compile(loss="binary_crossentropy")  # Wie gut ist die 1 und 0 Klassifizierung?
```

```python
model.fit(trainingL,
          trainingR,
          verbose=1,  # Outputs des Lernprozesses anzeigen
          epochs=10)  # Anzahl Lernprozesse
```

```python
model.fit(trainingL,
          trainingR,
          verbose=1,  # Outputs des Lernprozesses anzeigen
          epochs=40)  # Anzahl Lernprozesse
```

```python
model.save("Income.keras")
```

```python
predictions = model.predict(testL)
```

```python
predInt = (predictions > 0.5).astype(int).reshape(-1)
```

```python
pd.Series(predInt == testR.values).value_counts()
```

```python
predInt
```

```python
counts = pd.Series(predInt == testR.values).value_counts()
print(f"Korrekte Vorhersagen: {counts[True] / len(testR) * 100}%")
print(f"Falsche Vorhersagen: {counts[False] / len(testR) * 100}%")
```
