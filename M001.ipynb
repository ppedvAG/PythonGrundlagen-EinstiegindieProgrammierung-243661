## Shortcuts

- Shift + Enter: Zelle ausführen
- Strg + Enter: Zelle ausführen (aber nicht in die nächste Zelle)
- Enter: Zelle betreten
- Escape: Zelle verlassen (ohne diese auszuführen)

---

- X, C, V: Cut, Copy, Paste (für Zellen)
- A, B: Neue Zelle darunter/darüber einfügen
- D, D: Zelle löschen
- M, Y: Zwischen Markdown und Code umschalten
